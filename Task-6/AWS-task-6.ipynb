
### Step 1 - Create Training Data

```python
import cv2
import numpy as np

# Load HAAR face classifier
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Load functions
def face_extractor(img):
    # Function detects faces and returns the cropped face
    # If no face detected, it returns the input image
    
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray, 1.3, 5)
    
    if faces is ():
        return None
    
    # Crop all faces found
    for (x,y,w,h) in faces:
        cropped_face = img[y:y+h, x:x+w]

    return cropped_face

# Initialize Webcam
cap = cv2.VideoCapture(0)
count = 0

# Collect 100 samples of your face from webcam input for deepak
while True:

    ret, frame = cap.read()
    if face_extractor(frame) is not None:
        count += 1
        face = cv2.resize(face_extractor(frame), (200, 200))
        face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)

        # Save file in specified directory with unique name
        file_name_path = './faces/user-1/' + str(count) + '.jpg'
        cv2.imwrite(file_name_path, face)

        # Put count on images and display live count
        cv2.putText(face, str(count), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)
        cv2.imshow('Face Cropper', face)
        
    else:
        print("Face not found")
        pass

    if cv2.waitKey(1) == 13 or count == 100: #13 is the Enter Key
        break
        
cap.release()
cv2.destroyAllWindows()      
print("Successfully Collected Samples of Deepak")

n = input("Enter for collecting image for Gunu")

# Initialize Webcam for user-2
cap = cv2.VideoCapture(0)
count = 0
# Collect 100 samples of your face from webcam input for gunu
while True:

    ret, frame = cap.read()
    if face_extractor(frame) is not None:
        count += 1
        face1 = cv2.resize(face_extractor(frame), (200, 200))
        face1 = cv2.cvtColor(face1, cv2.COLOR_BGR2GRAY)

        # Save file in specified directory with unique name
        file_name_path1 = './faces/user-2/' + str(count) + '.jpg'
        cv2.imwrite(file_name_path1, face1)

        # Put count on images and display live count
        cv2.putText(face1, str(count), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)
        cv2.imshow('Face Cropper', face1)
        
    else:
        print("Face not found")
        pass

    if cv2.waitKey(1) == 13 or count == 100: #13 is the Enter Key
        break
        
cap.release()
cv2.destroyAllWindows()      
print("Successfully Collected Samples of Gunu")
```

### Step 2 - Train Model

```python
import cv2
import numpy as np
from os import listdir
from os.path import isfile, join

# Get the training data we previously made
data_path = './faces/user-1/'
onlyfiles = [f for f in listdir(data_path) if isfile(join(data_path, f))]

# Create arrays for training data and labels
Training_Data, Labels = [], []

# Open training images in our datapath
# Create a numpy array for training data
for i, files in enumerate(onlyfiles):
    image_path = data_path + onlyfiles[i]
    images = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    Training_Data.append(np.asarray(images, dtype=np.uint8))
    Labels.append(i)

# Create a numpy array for both training data and labels
Labels = np.asarray(Labels, dtype=np.int32)

# Initialize facial recognizer
# model = cv2.face.createLBPHFaceRecognizer()
# NOTE: For OpenCV 3.0 use cv2.face.createLBPHFaceRecognizer()
# pip install opencv-contrib-python
# model = cv2.createLBPHFaceRecognizer()

deepak_model  = cv2.face_LBPHFaceRecognizer.create()
# Let's train our model 
deepak_model.train(np.asarray(Training_Data), np.asarray(Labels))
print("Deepak Model trained sucessefully")


# Get the training data we previously made
data_path1 = './faces/user-2/'
onlyfiles1 = [f for f in listdir(data_path1) if isfile(join(data_path1, f))]

# Create arrays for training data and labels
Training_Data1, Labels1 = [], []

# Open training images in our datapath
# Create a numpy array for training data
for i, files in enumerate(onlyfiles1):
    image_path1 = data_path1 + onlyfiles1[i]
    images1 = cv2.imread(image_path1, cv2.IMREAD_GRAYSCALE)
    Training_Data1.append(np.asarray(images1, dtype=np.uint8))
    Labels1.append(i)

# Create a numpy array for both training data and labels
Labels1 = np.asarray(Labels1, dtype=np.int32)

# Initialize facial recognizer
# model = cv2.face.createLBPHFaceRecognizer()
# NOTE: For OpenCV 3.0 use cv2.face.createLBPHFaceRecognizer()
# pip install opencv-contrib-python
# model = cv2.createLBPHFaceRecognizer()

gunu_model  = cv2.face_LBPHFaceRecognizer.create()
# Let's train our model 
gunu_model.train(np.asarray(Training_Data1), np.asarray(Labels1))
print("Gunu Model trained sucessefully")
```
